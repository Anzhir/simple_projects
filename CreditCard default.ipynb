```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('1.3 UCI_Credit_Card.csv')
```

```python
categorical_cols=data[['SEX' ,'MARRIAGE', 'EDUCATION']]
```

```python
categorical_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
encoded=OneHotEncoder().fit_transform(categorical_cols).toarray()
```

```python
data.drop(['SEX' , 'MARRIAGE' , 'EDUCATION' ,'default.payment.next.month'] , axis=1 ,inplace=True )
```

```python
data.drop('ID' , axis=1 , inplace=True)
```

```python
data
```

```python
all_data=pd.concat([data , pd.DataFrame(encoded)] , axis=1)
all_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
all_data=scaler.fit_transform(all_data)
```

```python
pd.DataFrame(all_data)
```

```python
data=pd.read_csv('1.3 UCI_Credit_Card.csv')
labels=data['default.payment.next.month']
labels=np.array(labels)
```

```python
all_data=all_data.reshape((30000,33))
```

```python
all_data.shape
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(all_data , labels , test_size=0.3 )
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense , Dropout , BatchNormalization
model=Sequential()
model.add(Dense(5000 , activation='relu' , input_shape=(33,)))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(4000 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(3000 , activation='relu' ))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(2000 , activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.4))
model.add(Dense(1000 , activation='relu'))
model.add(Dropout(0.4))
model.add(BatchNormalization())
model.add(Dense(500 , activation='relu'))
model.add(Dropout(0.4))
model.add(BatchNormalization())
model.add(Dense(100 , activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(50 , activation='relu'))
model.add(Dense(1 , activation='sigmoid'))
model.compile(loss='binary_crossentropy' , optimizer='adam' , metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
early=EarlyStopping(monitor='val_loss' , patience=3)
model.fit(x_train , y_train , epochs=100 , callbacks=[early] , validation_split=0.2 , batch_size=32 )
```

```python
# summarize history for accuracy
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
model.evaluate(x_test , y_test)
```

```python
from keras.callbacks import EarlyStopping
early=EarlyStopping(monitor='val_loss' , patience=3)
model.fit(x_train , y_train , epochs=20 , callbacks=[early] , validation_split=0.2 , batch_size=32 )
```
